```python
# import libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
# Lodaing Database

df=pd.read_csv('creditcard.csv')
```

```python
# Exploratory Data Analysis

# shows haed of the dataset.

df.head(3)
```

```python
# Shows tails of the dataset

df.tail(3)
```

```python
df.info()
```

```python
# Checking shape of the data
df.shape
```

```python
# Chaecking Null data

df.isnull().sum()
```

```python
# Check Duplicate in this Dataset

df.duplicated().sum()
```

```python
# Drop Duplicate data

df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
# Replace Null value  place as 0.

df.fillna(0,inplace=True)
```

```python
df.isnull().sum().sum()
```

```python
# Get Overall Statistic about Dataset

df.describe()
```

```python
# Age data type changed float to Int .

df['age']=df['age'].astype(int)
```

```python
# Checking Columns

df.columns
```

```python
df.head(2)
```

```python
# Region wise cutomer

region_wise=df['Region'].unique()
print(region_wise)
```

```python
# number of customer each Region

number_of_customer=df['Region'].value_counts().reset_index()
number_of_customer=number_of_customer
print(number_of_customer)
```

```python
# Ploating A pie_Chart
plt.figure(figsize=(4,4))
plt.pie(number_of_customer['count'],labels=number_of_customer['Region'],autopct='%1.1f%%',explode=[0,0,0,0,0,0])
plt.title('Distribution of Customer')
plt.show()
```

```python
# Count number of Good_bad Customer
number_of_good_bad_customer=df['Good_Bad'].value_counts().reset_index()
print(number_of_good_bad_customer)
```

```python
number_of_good_bad_customer.info()
```

```python
df['Good_Bad'].value_counts()
```

```python
# Shows Good_bad Customer Through Bar .

plt.figure(figsize=(4,4))
sns.countplot(x='Good_Bad',data=df)
plt.xticks([1,0],['Good','Bad'])
plt.show()
```

```python
# Check Outliers for Debtratio.

df.columns
```

```python
df.Gender.value_counts()
```

```python
# Gender Distribution to the  Good_bad  variable.
plt.figure(figsize=(4,4))
sns.countplot(x='Gender',hue='Good_Bad',data=df)
plt.xticks([0,1],['Male','Female'])
plt.legend(labels=['Good','Bad'])
plt.show()
```

```python
df.columns
df.NumberOfOpenCreditLinesAndLoans.value_counts()
```

```python
# Check by age how many loan account open.

df.groupby('age')['NumberOfOpenCreditLinesAndLoans'].mean().sort_values(ascending=False)
```

```python
plt.figure(figsize=(10,10))
sns.barplot(y='age',x='NumberOfOpenCreditLinesAndLoans',data=df)
plt.title(' Age By loan Open ')
plt.show()
```
